Using elbow method to find the numbers of clusters in the datasets

x_1

```python
from sklearn.cluster import KMeans 
wc1 = []
for i in range(1,25):
    Kmeans = KMeans(n_clusters=i, init='k-means++' ,random_state=69)
    Kmeans.fit(x_1)
    wc1.append(Kmeans.inertia_)
```

```python
plt.plot(range(1,25),wc1)
plt.title('Elbow method')
plt.xlabel('The no of clusters')
plt.ylabel('WC1 values')
plt.show()
```
